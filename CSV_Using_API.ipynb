```python
import requests
import csv
from requests.api import head
```

```python
url='https://api.coincap.io/v2/assets'

headers={
    'Accept':'application/json',
    'Content-Type':'application/json'
}

response= requests.request("GET",url,headers=headers,data={})

js=response.json()

dt=[]
csvheader=['ID','RANK','SYMBOL','NAME','SUPPLY','PRICEUSD']

for x in js['data']:
  list=[x['id'],x['rank'],x['symbol'],x['name'],x['supply'],x['priceUsd']]
  dt.append(list)

with open('cryptoapi.csv','w',encoding='UTF8',newline='') as fl:
  writer=csv.writer(fl)


  writer.writerow(csvheader)
  writer.writerows(dt)

```
